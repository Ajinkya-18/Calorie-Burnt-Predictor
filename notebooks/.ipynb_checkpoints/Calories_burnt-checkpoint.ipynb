# Calories Burnt Prediction

## Data Loading and Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
calories_df = pd.read_csv('../data/calories.csv')
exercise_df = pd.read_csv('../data/exercise.csv')
```

```python
calories_df.head()
```

```python
exercise_df.head()
```

```python
df = pd.merge(exercise_df, calories_df, how='outer')
```

```python
df.head()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# Assuming height and weight units as cm and kg,
df['BMI'] = df['Weight'] / ((df['Height']/100)**2)
df.drop(['Weight', 'Height'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from collections import defaultdict

col_val_ranges = defaultdict(tuple)

for col in df.columns:
    if col == 'Gender':
        continue
    else:
        col_val_ranges[col] = (min(df[col]), max(df[col]))
```

## Feature Selection and Scaling

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Calories'], axis=1)
Y = df['Calories']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state = 42)
```

```python
x_train.shape, x_test.shape
```

```python
x_train.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.compose import ColumnTransformer

transformers = [('encoder', OrdinalEncoder(), ['Gender']), 
                ('scaler', StandardScaler(), ['Age', 'Duration', 'Heart_Rate', 'Body_Temp', 'BMI'])]

col_transformer = ColumnTransformer(transformers=transformers, remainder='passthrough', verbose=True, 
                                    sparse_threshold=0, n_jobs=5, verbose_feature_names_out=False)

col_transformer.set_output(transform='pandas')
```

```python
col_transformer.fit(x_train)
```

```python
# from joblib import dump
# with open('../models/column_transformer_fitted.joblib', 'wb') as f:
#     dump(col_transformer, f)
```

```python
x_train_scaled = col_transformer.transform(x_train)
x_test_scaled = col_transformer.transform(x_test)
```

```python
x_train_scaled.head()
```

```python
x_test_scaled.head()
```

```python
df_scaled = x_train_scaled.join(y_train, how='inner')
```

```python
plt.figure(figsize=(12, 12))
sb.heatmap(df_scaled.corr(), annot=True, fmt='.2f', cbar=True)
plt.show()
```

```python
x_train_scaled.drop(['Gender'], axis=1, inplace=True)
x_test_scaled.drop(['Gender'], axis=1, inplace=True)
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lin_svr = LinearSVR()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
knr = KNeighborsRegressor()
```

```python
x_train_scaled.columns, x_test_scaled.columns
```

```python
lr.fit(x_train_scaled, y_train)
lin_svr.fit(x_train_scaled, y_train)
dtr.fit(x_train_scaled, y_train)
rfr.fit(x_train_scaled, y_train)
knr.fit(x_train_scaled, y_train)
```

## Model Testing

```python
lr_preds = lr.predict(x_test_scaled)
lin_svr_preds = lin_svr.predict(x_test_scaled)
dtr_preds = dtr.predict(x_test_scaled)
rfr_preds = rfr.predict(x_test_scaled)
knr_preds = knr.predict(x_test_scaled)
```

```python
print("R2_Scores -->\n----------------------------------------------------------------------------------------------------")
print(f'LR: {round(r2_score(y_test, lr_preds), 5) * 100} %')
print(f'LinearSVR: {round(r2_score(y_test, lin_svr_preds), 5) * 100} %')
print(f'DTR: {round(r2_score(y_test, dtr_preds), 5) * 100} %')
print(f'RFR: {round(r2_score(y_test, rfr_preds), 5) * 100} %')
print(f'KNR: {round(r2_score(y_test, knr_preds), 5) * 100} %')
```

```python
print('Mean Squared Errors -->\n----------------------------------------------------------------------------------------------------')
print(f'LR: {round(mae(y_test, lr_preds), 5)}')
print(f'LinearSVR: {round(mae(y_test, lin_svr_preds), 5)}')
print(f'DTR: {round(mae(y_test, dtr_preds), 5)}')
print(f'RFR: {round(mae(y_test, rfr_preds), 5)}')
print(f'KNR: {round(mae(y_test, knr_preds), 5)}')
```

```python
fig, axs = plt.subplots(5, 1, figsize=(5, 20))
fig.suptitle('Model Prediction Performances')

axs[0].set_title('LR')
axs[0].scatter(y_test, lr_preds, color='black')
axs[0].set_xlabel('Actual Value (Y)')
axs[0].set_ylabel('Predicted Value (Y^)')


axs[1].set_title('LinearSVR')
axs[1].scatter(y_test, lin_svr_preds, color='green')
axs[1].set_xlabel('Actual Value (Y)')
axs[1].set_ylabel('Predicted Value (Y^)')

axs[2].set_title('DTR')
axs[2].scatter(y_test, dtr_preds, color='orange')
axs[2].set_xlabel('Actual Value (Y)')
axs[2].set_ylabel('Predicted Value (Y^)')

axs[3].set_title('RFR')
axs[3].scatter(y_test, rfr_preds, color='red')
axs[3].set_xlabel('Actual Value (Y)')
axs[3].set_ylabel('Predicted Value (Y^)')

axs[4].set_title('KNR')
axs[4].scatter(y_test, knr_preds, color='teal')
axs[4].set_xlabel('Actual Value (Y)')
axs[4].set_ylabel('Predicted Value (Y^)')

plt.tight_layout()
plt.savefig('../reports/model_performances.png')
```

```python
x_train_scaled.shape, x_test_scaled.shape
```

## SHAP Analysis

```python
import shap 
rfr_explainer = shap.TreeExplainer(rfr, x_train_scaled)
dtr_explainer = shap.TreeExplainer(dtr, x_train_scaled)
knr_explainer = shap.KernelExplainer(knr.predict, shap.sample(x_train_scaled, nsamples=5000))
lr_explainer = shap.LinearExplainer(lr, x_train_scaled)
svr_explainer = shap.KernelExplainer(lin_svr.predict, shap.sample(x_train_scaled, nsamples=500))
```

```python
rfr_shap_vals = rfr_explainer.shap_values(x_test_scaled, check_additivity=False)
dtr_shap_vals = dtr_explainer.shap_values(x_test_scaled, check_additivity=False)
lr_shap_vals = lr_explainer.shap_values(x_test_scaled)
knr_shap_vals = knr_explainer.shap_values(shap.sample(x_test_scaled, nsamples=100))
svr_shap_vals = svr_explainer.shap_values(shap.sample(x_test_scaled, nsamples=100))
```

```python
plt.title('RFR SHAP Values')
shap.summary_plot(rfr_shap_vals, x_test_scaled, plot_type='bar')
plt.savefig('../reports/RFR_SHAP_Values.png')
```

```python
plt.title('DTR SHAP Values')
shap.summary_plot(dtr_shap_vals, x_test_scaled, plot_type='bar')
plt.savefig('../reports/DTR_SHAP_Values.png')
```

```python
plt.title('LR SHAP Values')
shap.summary_plot(lr_shap_vals, x_test_scaled, plot_type='bar')
plt.savefig('../reports/LR_SHAP_Values.png')
```

```python
plt.title('KNR SHAP Values')
shap.summary_plot(knr_shap_vals, shap.sample(x_test_scaled, nsamples=100), plot_type='bar')
plt.savefig('../reports/KNR_SHAP_Values.png')
```

```python
plt.title('Linear SVR SHAP Values')
shap.summary_plot(svr_shap_vals, shap.sample(x_test_scaled, nsamples=100), plot_type='bar')
plt.savefig('../reports/RFR_SHAP_Values.png')
```
